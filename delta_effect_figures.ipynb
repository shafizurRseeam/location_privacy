```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

def plot_l2_distance_effect(base_directory, dataset_name):
    # Base directory for the perturbed data
    base_directory_input = os.path.join(base_directory, dataset_name, 'perturbed')
    
    # Epsilon values and Delta configurations
    epsilon_values = [0.1, 1, 5]
    deltas = [5, 10, 20]
    l2_distances = []

    for delta in deltas:
        x_values = []
        for epsilon in epsilon_values:
            file_name = f'merged_our_bl_50_delta_{delta}_{epsilon}.csv'
            file_path = os.path.join(base_directory_input, f'our_bl_50_delta_{delta}', file_name)
            df = pd.read_csv(file_path)
            mean_value = df['l2_distance'].mean()
            x_values.append(mean_value)
        l2_distances.append(x_values)

    # Plot setup
    fig, ax = plt.subplots(figsize=(10, 6))
    colors = ['#1F77B4', '#FF7F0E', '#2CA02C']
    width = 0.2
    delta_gap = 0.1
    group_width = len(epsilon_values) * width + delta_gap
    x = np.arange(len(deltas)) * group_width

    # Plotting bars
    for delta_index, delta in enumerate(deltas):
        for bar_index, epsilon in enumerate(epsilon_values):
            bar_position = x[delta_index] + bar_index * width
            bar_height = l2_distances[delta_index][bar_index]
            ax.bar(bar_position, bar_height, width, color=colors[delta_index], edgecolor='black')

            # Text above bars
            text_y_position = bar_height 
            ax.text(bar_position-width/2, text_y_position, f'$\epsilon$ = {epsilon}',
                    ha='left', va='bottom', fontsize=20, rotation=45)

    # Customization
    ax.set_xlabel('$\delta$', fontsize=32)
    ax.set_ylabel('Average $L_2$ Distance', fontsize=28)
    ax.set_xticks(x + width * (len(epsilon_values) - 1) / 2)
    ax.set_xticklabels(deltas)
    plt.tick_params(axis='both', which='major', labelsize=28)
    plt.ylim(0, max(l2_distances[2]) * 1.3)

    # Save and display
    save_dir_png = os.path.join(base_directory, dataset_name, 'figure', 'png')
    save_dir_svg = os.path.join(base_directory, dataset_name, 'figure', 'svg')
    os.makedirs(save_dir_png, exist_ok=True)
    os.makedirs(save_dir_svg, exist_ok=True)
    pathpng = os.path.join(save_dir_png, f'{dataset_name}_l2_delta_effect.png')
    pathsvg = os.path.join(save_dir_svg, f'{dataset_name}_l2_delta_effect.svg')
    plt.savefig(pathpng, dpi=300, bbox_inches='tight')
    plt.savefig(pathsvg, format='svg', bbox_inches='tight')
    plt.show()

# Example usage:
base_directory = "C:\\Users\\ss6365\\Desktop\\location_privacy_final"
dataset_name = 'geolife'  # Change 'uci' to any other directory name as needed
plot_l2_distance_effect(base_directory, dataset_name)
```
