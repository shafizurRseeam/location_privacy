```python
import math
import random
import json
import pandas as pd
from haversine import haversine, Unit

# =============================================================================
# COMMON UTILITIES
# =============================================================================

OVERLAY_RADIUS = 100         # in meters: radius of true/perturbed circle
POKEMON_MAX_RADIUS = 1000    # in meters: radius around fixed center to generate Pokémon
POKEMON_SPACING = 100        # in meters: spacing for generating Pokémon positions
#sparse pokemons
def load_noise_pairs(filename):
    with open(filename, 'r') as f:
        noise_data = json.load(f)
    return [(pair[0], pair[1]) for pair in noise_data]

def distance_meters(lat1, lon1, lat2, lon2):
    avg_lat = math.radians((lat1 + lat2) / 2)
    dx = (lon2 - lon1) * (111320 * math.cos(avg_lat))
    dy = (lat2 - lat1) * 111320
    return math.sqrt(dx * dx + dy * dy)

def add_noise_to_location(lat, lon, noise_x, noise_y):
    delta_lat = noise_y / 111320.0
    delta_lon = noise_x / (111320.0 * math.cos(math.radians(lat)))
    return lat + delta_lat, lon + delta_lon

def generate_uniform_pokemon_positions(center, max_radius, spacing):
    center_lat, center_lon = center
    positions = []
    meter_to_deg_lat = 1.0 / 111320.0
    meter_to_deg_lon = 1.0 / (111320.0 * math.cos(math.radians(center_lat)))
    radius_deg_lat = max_radius * meter_to_deg_lat
    radius_deg_lon = max_radius * meter_to_deg_lon
    min_lat = center_lat - radius_deg_lat
    max_lat = center_lat + radius_deg_lat
    min_lon = center_lon - radius_deg_lon
    max_lon = center_lon + radius_deg_lon
    spacing_deg_lat = spacing * meter_to_deg_lat
    spacing_deg_lon = spacing * meter_to_deg_lon

    lat = min_lat
    while lat <= max_lat:
        lon = min_lon
        while lon <= max_lon:
            if distance_meters(center_lat, center_lon, lat, lon) <= max_radius:
                positions.append((lat, lon))
            lon += spacing_deg_lon
        lat += spacing_deg_lat
    return positions

# =============================================================================
# HELPER FUNCTIONS FOR GEOGRAPHIC CONVERSIONS (for intermediate mechanism)
# =============================================================================

def latitude_longitude_to_y_x(latitude, longitude):
    EARTH_RADIUS = 6371000  # in meters
    x = EARTH_RADIUS * math.radians(longitude)
    y = EARTH_RADIUS * math.radians(latitude)
    return y, x

def y_x_to_latitude_longitude(y, x):
    EARTH_RADIUS = 6371000
    longitude = math.degrees(x / EARTH_RADIUS)
    latitude = math.degrees(y / EARTH_RADIUS)
    return latitude, longitude

def get_direction(loc1, loc2):
    lat1, lon1 = map(math.radians, loc1)
    lat2, lon2 = map(math.radians, loc2)
    dLon = lon2 - lon1
    y = math.sin(dLon) * math.cos(lat2)
    x = math.cos(lat1)*math.sin(lat2) - math.sin(lat1)*math.cos(lat2)*math.cos(dLon)
    return math.atan2(y, x)

def inverse_haversine(start, distance, bearing, unit=Unit.METERS):
    R = 6371000  # meters
    lat1 = math.radians(start[0])
    lon1 = math.radians(start[1])
    d_div_R = distance / R
    lat2 = math.asin(math.sin(lat1)*math.cos(d_div_R) +
                     math.cos(lat1)*math.sin(d_div_R)*math.cos(bearing))
    lon2 = lon1 + math.atan2(math.sin(bearing)*math.sin(d_div_R)*math.cos(lat1),
                             math.cos(d_div_R)-math.sin(lat1)*math.sin(lat2))
    return math.degrees(lat2), math.degrees(lon2)

# =============================================================================
# MECHANISM FUNCTIONS
# =============================================================================

# Direct Noise Addition (Laplace and Staircase)
def planar_laplace_mechanism_point(location, noise_laplace):
    lat, lon = location
    noise_x, noise_y = random.choice(noise_laplace)
    return add_noise_to_location(lat, lon, noise_x, noise_y)

def planar_staircase_mechanism_point(location, noise_staircase):
    lat, lon = location
    y, x = latitude_longitude_to_y_x(lat, lon)
    noise_x, noise_y = random.choice(noise_staircase)
    perturbed_x = x + noise_x
    perturbed_y = y + noise_y
    return y_x_to_latitude_longitude(perturbed_y, perturbed_x)

# Staircase with Intermediate Mechanism
def bounded_planar_staircase_mechanism_point(location, noise_staircase_bounded):
    lat, lon = location
    y, x = latitude_longitude_to_y_x(lat, lon)
    noise_x, noise_y = random.choice(noise_staircase_bounded)
    perturbed_x = x + noise_x
    perturbed_y = y + noise_y 
    return y_x_to_latitude_longitude(perturbed_y, perturbed_x)

def intermediate(df, noise_staircase_bounded):
    df = df.copy()
    n = len(df)
    df['intermediate_lat'] = None
    df['intermediate_lon'] = None

    # First location
    true_loc = (df.loc[0, 'latitude'], df.loc[0, 'longitude'])
    inter_loc = bounded_planar_staircase_mechanism_point(true_loc, noise_staircase_bounded)
    df.at[0, 'intermediate_lat'] = inter_loc[0]
    df.at[0, 'intermediate_lon'] = inter_loc[1]

    # Subsequent locations
    for i in range(1, n):
        true_loc = (df.loc[i, 'latitude'], df.loc[i, 'longitude'])
        last_true = (df.loc[i-1, 'latitude'], df.loc[i-1, 'longitude'])
        d = haversine(true_loc, last_true, unit=Unit.METERS)
        bearing = get_direction(last_true, true_loc)
        prev_intermediate = (df.loc[i-1, 'intermediate_lat'], df.loc[i-1, 'intermediate_lon'])
        inter_loc = inverse_haversine(prev_intermediate, d, bearing, Unit.METERS)
        df.at[i, 'intermediate_lat'] = inter_loc[0]
        df.at[i, 'intermediate_lon'] = inter_loc[1]
    return df

def only_reported_locations(df, noise_staircase, delta):
    df = df.copy()
    df['perturbed_latitude'] = None
    df['perturbed_longitude'] = None

    reported = planar_staircase_mechanism_point((df.loc[0, 'intermediate_lat'], df.loc[0, 'intermediate_lon']),
                                                 noise_staircase)
    df.at[0, 'perturbed_latitude'] = reported[0]
    df.at[0, 'perturbed_longitude'] = reported[1]
    current_focus = (df.loc[0, 'latitude'], df.loc[0, 'longitude'])
    current_reported = reported

    for i in range(1, len(df)):
        true_loc = (df.loc[i, 'latitude'], df.loc[i, 'longitude'])
        if haversine(true_loc, current_focus, unit=Unit.METERS) < delta:
            df.at[i, 'perturbed_latitude'] = current_reported[0]
            df.at[i, 'perturbed_longitude'] = current_reported[1]
        else:
            reported = planar_staircase_mechanism_point((df.loc[i, 'intermediate_lat'], df.loc[i, 'intermediate_lon']),
                                                         noise_staircase)
            df.at[i, 'perturbed_latitude'] = reported[0]
            df.at[i, 'perturbed_longitude'] = reported[1]
            current_focus = true_loc
            current_reported = reported
    return df

# =============================================================================
# SIMULATION FUNCTIONS
# =============================================================================

def simulate_show_rate_for_point(true_location, noise_pairs, pokemon_positions, iterations=1, mechanism='laplace'):
    true_lat, true_lon = true_location
    total_true = 0
    total_intersection = 0

    for _ in range(iterations):
        if mechanism == 'laplace':
            perturbed = add_noise_to_location(true_lat, true_lon, *random.choice(noise_pairs))
        elif mechanism == 'staircase':
            y, x = latitude_longitude_to_y_x(true_lat, true_lon)
            noise_x, noise_y = random.choice(noise_pairs)
            perturbed = y_x_to_latitude_longitude(y + noise_y, x + noise_x)
        else:
            raise ValueError("Unknown mechanism type")
        for p in pokemon_positions:
            if distance_meters(true_lat, true_lon, p[0], p[1]) <= OVERLAY_RADIUS:
                total_true += 1
                if distance_meters(perturbed[0], perturbed[1], p[0], p[1]) <= OVERLAY_RADIUS:
                    total_intersection += 1
    return total_true, total_intersection

def simulate_for_users(noise_filename, users_trajectories, fixed_pokemon_center, iterations_per_point=1, mechanism='laplace'):
    noise_pairs = load_noise_pairs(noise_filename)
    pokemon_positions = generate_uniform_pokemon_positions(fixed_pokemon_center, POKEMON_MAX_RADIUS, POKEMON_SPACING)
    overall_true = 0
    overall_intersection = 0

    for trajectory in users_trajectories:
        user_true = 0
        user_intersection = 0
        for point in trajectory:
            t_true, t_intersection = simulate_show_rate_for_point(point, noise_pairs, pokemon_positions,
                                                                  iterations=iterations_per_point, mechanism=mechanism)
            user_true += t_true
            user_intersection += t_intersection
        overall_true += user_true
        overall_intersection += user_intersection

    overall_average = (overall_intersection / overall_true) * 100 if overall_true > 0 else 0
    return overall_average

def simulate_show_rate_for_trajectory_intermediate(trajectory, bounded_noise, unbounded_noise, delta, pokemon_positions, iterations=1):
    total_true = 0
    total_intersection = 0

    for _ in range(iterations):
        df = pd.DataFrame(trajectory, columns=['latitude', 'longitude'])
        df = intermediate(df, bounded_noise)
        df = only_reported_locations(df, unbounded_noise, delta)
        for idx, row in df.iterrows():
            true_loc = (row['latitude'], row['longitude'])
            rep_loc = (row['perturbed_latitude'], row['perturbed_longitude'])
            for p in pokemon_positions:
                if haversine(true_loc, p, unit=Unit.METERS) <= OVERLAY_RADIUS:
                    total_true += 1
                    if haversine(rep_loc, p, unit=Unit.METERS) <= OVERLAY_RADIUS:
                        total_intersection += 1
    return total_true, total_intersection

def simulate_for_users_intermediate(bounded_noise_file, unbounded_noise_file, users_trajectories, fixed_pokemon_center, delta, iterations_per_point=1):
    bounded_noise = load_noise_pairs(bounded_noise_file)
    unbounded_noise = load_noise_pairs(unbounded_noise_file)
    pokemon_positions = generate_uniform_pokemon_positions(fixed_pokemon_center, POKEMON_MAX_RADIUS, POKEMON_SPACING)
    overall_true = 0
    overall_intersection = 0

    for idx, trajectory in enumerate(users_trajectories):
        #print(f"User {idx+1} trajectory:")
        t_true, t_intersection = simulate_show_rate_for_trajectory_intermediate(trajectory, bounded_noise, unbounded_noise, delta, pokemon_positions, iterations=iterations_per_point)
        user_show_rate = (t_intersection / t_true) * 100 if t_true > 0 else 0
        #print(f"  Overall average show rate for user {idx+1}: {user_show_rate:.2f}%\n")
        overall_true += t_true
        overall_intersection += t_intersection

    overall_average = (overall_intersection / overall_true) * 100 if overall_true > 0 else 0
    return overall_average


```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

def plot_score_vs_epsilon(data_dir, figure_dir, dataset_name, scores_dict, epsilon_values):
    """
    Plots and saves Score vs Epsilon graph following the same style as generate_distance_plots.

    Parameters:
    - data_dir: Base data directory.
    - figure_dir: Base directory for saving figures.
    - dataset_name: Name of the dataset.
    - scores_dict: Dictionary with keys as method names and values as lists of scores.
                   Example: {'laplace': [...], 'staircase': [...]}
    - epsilon_values: List of epsilon values corresponding to the scores (e.g., [0.1, 0.2, 0.3])
    """
    plt.figure(figsize=(10, 6))

    plot_attributes = {
        'laplace': {'label': r'$\mathtt{PL}_ϵ$', 'color': '#1F77B4', 'marker': 'o'},
        'staircase': {'label': r'$\mathtt{PS}_ϵ$', 'color': '#FF7F0E', 'marker': 'd'},
        'our_bl_50_delta_5': {'label':  r'$\mathtt{PS}_ϵ\mathtt{I}$', 'color': '#2CA02C', 'marker': 'X'}
    }

    for method, scores in scores_dict.items():
        if method in plot_attributes:
            attr = plot_attributes[method]
            plt.plot(
                epsilon_values, scores,
                label=attr['label'],
                color=attr['color'],
                marker=attr['marker'],
                linestyle='solid',
                linewidth=2,
                markersize=14
            )

    #plt.xlabel('$\epsilon$', fontsize=34)
    #plt.ylabel('Score', fontsize=30)
    plt.xscale('log')
    plt.xticks(epsilon_values, labels=[str(e) for e in epsilon_values])
    plt.tick_params(axis='both', which='major', labelsize=30)
    #plt.grid(True)
    plt.legend(fontsize=30, frameon=True, shadow=True).get_frame().set_edgecolor('black')
    # ax = plt.gca()
    # ax.yaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f'))
    # Save the plot
    figure_base_dir = os.path.join(figure_dir, dataset_name, 'figure')
    png_dir = os.path.join(figure_base_dir, 'png')
    svg_dir = os.path.join(figure_base_dir, 'svg')
    os.makedirs(png_dir, exist_ok=True)
    os.makedirs(svg_dir, exist_ok=True)

    filename = f'{dataset_name}_score_vs_epsilon'
    plt.savefig(os.path.join(png_dir, f'{filename}.png'), dpi=300, bbox_inches='tight')
    plt.savefig(os.path.join(svg_dir, f'{filename}.svg'), format='svg', bbox_inches='tight')

    plt.show()

# Example usage
data_dir = r"C:\Users\ss6365\Desktop\ndss\figures"
figure_dir = data_dir
dataset_name = 'pokemon_sparse'
# Scores for each method

scores = {
    'laplace': [results_eps_0_1[0], results_eps_0_2[0], results_eps_0_5[0], results_eps_1[0], results_eps_2[0]],
    'staircase': [results_eps_0_1[1], results_eps_0_2[1], results_eps_0_5[1], results_eps_1[1], results_eps_2[1]],
    'our_bl_50_delta_5':[results_eps_0_1[2], results_eps_0_2[2], results_eps_0_5[2], results_eps_1[2], results_eps_2[2]]
}


# Corresponding epsilon values
epsilons = [0.1, 0.2, 0.5, 1, 2]

plot_score_vs_epsilon(data_dir, figure_dir, dataset_name, scores, epsilons)
```
